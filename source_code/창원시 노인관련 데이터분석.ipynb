```python
gyeong_ro_dang = pd.read_csv('../data/경상남도_창원시_경로당현황.csv', encoding='euc-kr')
mael = pd.read_csv('../data/경상남도_창원시_마을회관현황.csv', encoding='euc-kr')
elder_pop = pd.read_excel('../data/노인인구현황.xlsx', header=1)
```

```python
elder = pd.merge(gyeong_ro_dang, mael, on='읍면동', how='outer')
elder
```

```python
elder['노인복지시설수'] = elder['경로당수'] + elder['마을회관수']
elder
```

```python
np.corrcoef(elder['노인복지시설수'], elder['노인인구수'])
```

```python
elder['노인인구수'].sort_values().plot(kind='barh', grid=True, figsize=(10,20))
plt.show()
```

```python
fp1 = np.polyfit(elder['노인인구수'], elder['노인복지시설수'],1)
f1 = np.poly1d(fp1)
fx = np.linspace(0, 7000, 10)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(elder['노인인구수'], elder['노인복지시설수'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('노인인구수')
plt.ylabel('복지시설수')
plt.grid()
plt.show()
```

```python
elder['오차'] = np.abs(elder['노인복지시설수'] - f1(elder['노인인구수']))
elder_sort = elder.sort_values(by='오차', ascending=False)

plt.figure(figsize=(10,10))
plt.scatter(elder['노인인구수'], elder['노인복지시설수'], c=elder['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(elder_sort['노인인구수'][n]*1.02, elder_sort['노인복지시설수'][n]*0.98,
            elder_sort.index[n], fontsize=8)

plt.xlabel('노인인구수')
plt.ylabel('복지시설수')
plt.colorbar()
plt.grid()
plt.show()
```

```python
import seaborn as sns
tmp_max = elder['노인인구수'].max()
elder['노인인구수(정규화)'] = elder['노인인구수'] / tmp_max * 100
tmp_max = elder['경로당수'].max()
elder['경로당수(정규화)'] = elder['경로당수'] / tmp_max * 100
tmp_max = elder['마을회관수'].max()
elder['마을회관수(정규화)'] = elder['마을회관수'] / tmp_max * 100
tmp_max = elder['노인복지시설수'].max()
elder['노인복지시설수(정규화)'] = elder['노인복지시설수'] / tmp_max * 100

target_col = ['경로당수(정규화)', '마을회관수(정규화)', '노인인구수(정규화)', '노인복지시설수(정규화)']

elder_sort = elder.sort_values(by='노인복지시설수', ascending=False)

plt.figure(figsize=(10,20))
sns.heatmap(elder_sort[target_col], annot=True, fmt='f', linewidths=.5)
plt.title('읍면동노인복지시설 비율')
plt.show()
```
